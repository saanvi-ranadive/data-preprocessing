```python
# 1

import pandas as pd

def read_gct_to_df(file_path, summary=False):
    """
    Reads data from GenePattern GCT file into pandas DataFrame.
    
    Parameters:
        file_path (str): File path of GCT file.
        summary (bool, optional): Optional parameter with default value False. If set to True, prints number of rows and columns imported from the file.
        
    Returns:
        df (pandas.DataFrame): DataFrame containing data from GCT file
    """
    # Open file in read mode. 
    with open(file_path, "r") as f:
        
        # Skip first line.
        next(f)
        
        # Save second line as number of rows and columns.
        rows_and_columns = f.readline().strip()
        
        # Read file into DataFrame, argument specifies that columns are separated by a tab character ("\t")
        df = pd.read_csv(f, sep="\t")
        
        # Set "Name" and "Description" to be indices of the DataFrame.
        df = df.set_index(["Name", "Description"])
    
    # Print number of rows and columns if summary equals True.
    if summary == True:
        print(rows_and_columns)
    
    return df
```

```python
# 3

import matplotlib.pyplot as plt

# Read data from the GCT file into a DataFrame.
df = read_gct_to_df("BRCA_large_20783x40.gct")

# Define list of samples to be plotted as histograms.
samples = ["A7-A0DB-normal", "A7-A13E-normal", "BH-A0B3-primary", "BH-A0B5-primary"]

# Create a 2x2 grid of subplots.
fig, axes = plt.subplots(2, 2, figsize=(16, 16))

# Plot histogram for each sample.
for i in range(len(samples)):
    sample_clean = df[samples[i]].dropna() # Drop NA entries from sample column.
    axes[(i//2, i%2)].hist(sample_clean, bins=15, range=(0, 10000)) # Plot histogram on the corresponding subplot.
    axes[(i//2, i%2)].set_title("Histogram of " + str(samples[i])) # Set histogram title.

plt.show()
```

```python
# 4

# Read data from the GCT file into a DataFrame.
df = read_gct_to_df("BRCA_large_20783x40.gct")

# Initialize DataFrame to hold summary statistics for each sample.
summary_stats_df = pd.DataFrame(columns=["Sample Name", "Mean", "Median", "Standard Deviation"])

# Define list of samples to be plotted as histograms.
samples = ["A7-A0DB-normal", "A7-A13E-normal", "BH-A0B3-primary", "BH-A0B5-primary"]

# Add summary statistics for each sample to summary_stats_df.
for sample in samples:
    sample_clean = df[sample].dropna() # Drop NA entries from sample column.
    mean = sample_clean.mean() # Find mean.
    median = sample_clean.median() # Find median.
    standard_dev = sample_clean.std() # Find standard deviation.
    summary_stats_df.loc[len(summary_stats_df)] = [sample, mean, median, standard_dev] # Add row to summary_stats_df.

summary_stats_df
```
